# Population Parameters

**Population:**
A population refers to a comprehensive collection of elements, whether individuals, items, or entities, sharing common characteristics, from which data is sought.

**Parameter:**
A parameter is a numerical characteristic of a population, often represented by symbols such as μ (mu) for the population mean or p for the population proportion.

Illustrative Examples:

- Our inquiry may involve determining the mean height of all adult males in a given region. Here, the population encompasses all adult males, and the parameter is denoted by μ.

- Alternatively, our focus could be on ascertaining the proportion of defective products produced by a manufacturing plant. The population, in this case, consists of all products from the plant, and the relevant parameter is represented by p.

In practice, the exact value of a population parameter is frequently unknown or impractical to ascertain. Estimation becomes imperative, with samples and statistics serving as essential tools in this process.

```python
import numpy as np

# Generate a synthetic population of heights of adult males in a given region
population_heights = np.random.normal(loc=175, scale=5, size=10000)

# Calculate the population mean (μ)
population_mean = np.mean(population_heights)

# Display the population mean
print(f"Population Mean (μ): {population_mean}")

# Generate a synthetic population of proportions of defective products in a manufacturing plant
population_defect_proportions = np.random.uniform(low=0, high=0.1, size=10000)

# Calculate the population proportion (p)
population_defect_proportion = np.mean(population_defect_proportions)

# Display the population proportion
print(f"Population Proportion (p): {population_defect_proportion}")

# Simulate the process of estimating the population parameters from samples
sample_size = 100
num_samples = 1000

sample_means = np.zeros(num_samples)
sample_proportions = np.zeros(num_samples)

for i in range(num_samples):
    # Take a random sample from the population
    sample = np.random.choice(population_heights, size=sample_size)

    # Calculate sample mean and proportion
    sample_means[i] = np.mean(sample)

    # For illustration, let's use the proportion example again
    sample_defect_proportion = np.mean(np.random.choice(population_defect_proportions, size=sample_size))
    sample_proportions[i] = sample_defect_proportion

# Calculate the average of sample means and sample proportions as estimators
estimated_population_mean = np.mean(sample_means)
estimated_population_proportion = np.mean(sample_proportions)

# Display the estimated population parameters
print(f"\nEstimated Population Mean from Samples: {estimated_population_mean}")
print(f"Estimated Population Proportion from Samples: {estimated_population_proportion}")
```

# Sample Statistics

**Sample:**
A sample is a representative subset drawn from a population, intended to reflect its characteristics.

**Statistic:**
A statistic is a numerical summary, such as an average or percentage, describing key features of a sample.

The sample mean, denoted as X¯, and the sample proportion, denoted as p̂, represent distinct sample statistics. For instance:

- We may utilize X¯, the average weight of a random sample of 100 middle-aged female Americans, to estimate μ, the average weight of all middle-aged female Americans.
  
- Alternatively, p̂, the proportion in a random sample of 1000 likely American voters approving of the president's job performance, can be used to estimate p, the proportion of all likely American voters with a positive view of the president's job performance.

Due to the manageable size of samples, we can precisely determine the value of any statistic. The known value of the sample statistic is then employed to make inferences about the unknown value of the population parameter.

The techniques employed to succinctly portray data and capture its essential information fall under descriptive statistics. For instance, the sample mean serves as a descriptive statistic. Conversely, the process of drawing conclusions about the population from data is known as inferential statistics.

```python
import numpy as np

# Generating a synthetic population (e.g., heights of all adults)
population = np.random.normal(loc=170, scale=10, size=10000)

# Taking a sample from the population
sample_size = 100
sample = np.random.choice(population, size=sample_size)

# Calculating sample statistics
sample_mean = np.mean(sample)
sample_proportion = np.sum(sample > 175) / sample_size  # Example of a proportion

# Displaying results
print(f"Population Mean: {np.mean(population)}")
print(f"Sample Mean: {sample_mean}")

# Performing inferential statistics (estimation)
estimated_population_mean = np.mean(sample)
print(f"Estimated Population Mean from Sample: {estimated_population_mean}")

# Descriptive statistics (capturing the essence of the data)
print(f"Descriptive Statistics of Sample:\n{np.round(np.percentile(sample, [25, 50, 75]), 2)}")

# Inferential statistics (drawing conclusions)
# In this simplified example, let's assume a hypothesis testing scenario.
# Null hypothesis: Population mean is equal to a certain value (e.g., 170)
# Significance level: 0.05
from scipy.stats import ttest_1samp

t_statistic, p_value = ttest_1samp(sample, 170)
print(f"\nHypothesis Testing Results:")
print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")
if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

# Data Collection

Initiating the process of statistical analysis entails the critical step of data collection. The overarching objective of statistical methods is to draw meaningful inferences about a population based on information derived from a carefully selected sample. The methodology employed in data collection plays a pivotal role, as the representativeness of the sample directly influences the validity of inferences drawn about the entire population.

Several widely utilized methods for data collection include:

1. **Personal Interview:**
   - Prospective respondents typically exhibit a higher willingness to engage when queried by an interviewer. However, responses may be susceptible to influence from the interviewer, introducing a potential source of bias.

2. **Telephone Interview:**
   - A cost-effective approach, albeit necessitating brevity due to respondent impatience. The challenge lies in balancing efficiency with obtaining comprehensive data.

3. **Self-Administered Questionnaires:**
   - An economical option, albeit accompanied by a lower response rate. This method may introduce bias as respondents self-select, potentially resulting in a non-representative sample.

4. **Direct Observation:**
   - Particularly applicable when the attribute of interest can be directly measured within the sample. This method offers objectivity in assessing certain quantitative aspects.

5. **Web-Based Survey:**
   - Tailored to populations with internet access, a web-based survey offers a modern approach. However, the limitation arises from its exclusion of individuals without web connectivity.

Each method possesses unique advantages and challenges, and the selection of an appropriate approach hinges on the research objectives and characteristics of the target population.

```python
import numpy as np
import pandas as pd

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic data representing a population
population_size = 10000
population_data = pd.DataFrame({
    'Age': np.random.normal(loc=30, scale=5, size=population_size),
    'Income': np.random.normal(loc=50000, scale=10000, size=population_size),
    'InternetUsage': np.random.uniform(low=0, high=1, size=population_size)
})

# Simulate the process of selecting a sample
sample_size = 500
sample_indices = np.random.choice(population_data.index, size=sample_size, replace=False)
sample_data = population_data.loc[sample_indices]

# Display the synthetic population and sample data
print("Synthetic Population Data:")
print(population_data.head())

print("\nSynthetic Sample Data:")
print(sample_data.head())

# Simulate different data collection methods
# For simplicity, we'll use a random attribute 'response' to indicate willingness to participate
population_data['Response'] = np.random.choice([0, 1], size=population_size, p=[0.2, 0.8])
sample_data['Response'] = np.random.choice([0, 1], size=sample_size, p=[0.2, 0.8])

# Display the response rates in the population and sample
population_response_rate = population_data['Response'].mean()
sample_response_rate = sample_data['Response'].mean()

print(f"\nPopulation Response Rate: {population_response_rate:.2%}")
print(f"Sample Response Rate: {sample_response_rate:.2%}")
```

# Types of Bias in Data Collection

Data collection inherently carries the risk of introducing bias, where the characteristics of the sample may deviate from those of the intended population. Several types of bias are particularly noteworthy:

1. **Non-Response Bias:**
   - Occurs when a substantial proportion of the selected individuals decline to respond or participate in the study. The non-participation of a significant segment may distort the representativeness of the sample.

2. **Response Bias:**
   - Arises when study participants provide inaccurate responses, either by not disclosing the truth or by tailoring their answers to align with perceived expectations of the researcher. For instance, respondents might be inclined to deny engaging in certain behaviors to conform to societal norms.

3. **Selection Bias:**
   - Materializes when the chosen sample fails to accurately mirror the characteristics of the target population. For example, if the research focus is the attitudes of female students toward campus safety, but the sample includes both males and females, the selection bias emerges due to the inclusion of subjects outside the specified population of interest.

Understanding and mitigating these biases is imperative for maintaining the integrity and validity of statistical inferences drawn from the collected data.

```python
import numpy as np
import pandas as pd

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic data representing a population
population_size = 10000
population_data = pd.DataFrame({
    'Gender': np.random.choice(['Male', 'Female'], size=population_size),
    'CampusSafetyAttitude': np.random.choice(['Positive', 'Neutral', 'Negative'], size=population_size),
})

# Simulate response bias where some individuals may not respond truthfully
biased_response_indices = np.random.choice(population_data.index, size=int(0.1 * population_size), replace=False)
population_data.loc[biased_response_indices, 'CampusSafetyAttitude'] = 'Positive'

# Display the synthetic population data
print("Synthetic Population Data:")
print(population_data.head())

# Simulate a sample selection focusing on female students
sample_size = 500
sample_indices = population_data[population_data['Gender'] == 'Female'].sample(sample_size, replace=False).index
sample_data = population_data.loc[sample_indices]

# Display the synthetic sample data
print("\nSynthetic Sample Data:")
print(sample_data.head())
```

# Data Collection Methods

The process of data collection involves deliberate decisions on how to obtain information and select observations or measurements for a study. Two overarching methods guide this process: non-probability methods and probability methods.

**Non-Probability Methods:**

1. **Convenience Sampling (Haphazard):**
   - Involves collecting data from subjects readily available or conveniently obtained, often without a systematic approach.
   - Example: Surveying students as they pass by in the university's student union building.

2. **Gathering Volunteers:**
   - Involves collecting data from subjects who willingly volunteer to provide information.
   - Example: Utilizing advertisements in magazines or on websites, inviting individuals to complete a form or participate in a study.

**Probability Methods:**

1. **Simple Random Sample:**
   - Entails making selections from a population where each subject has an equal and independent chance of being chosen.
   
2. **Stratified Random Sample:**
   - Begins with identifying the population of interest and then divides it into strata or groups based on specific characteristics (e.g., sex, geographic region). A simple random sample is then taken from each stratum.

3. **Cluster Sample:**
   - Involves selecting a random cluster of subjects from the population of interest. For instance, estimating the average salary for faculty members at a university might involve taking a simple random sample of departments and determining the salary of each faculty member within the sampled department.

Both non-probability and probability methods have their advantages and disadvantages. Non-probability methods are often more straightforward and cost-effective but may lead to non-representative samples. Probability sampling methods, on the other hand, allow for making inferences and generalizations about the population based on a representative sample obtained through random selection.

```python
import numpy as np
import pandas as pd

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic data representing a population of students
population_size = 1000
population_data = pd.DataFrame({
    'StudentID': np.arange(1, population_size + 1),
    'Age': np.random.randint(18, 25, size=population_size),
    'GPA': np.random.uniform(2.0, 4.0, size=population_size)
})

# Display the synthetic population data
print("Synthetic Population Data:")
print(population_data.head())

# Perform simple random sampling
sample_size = 100
sample_indices = np.random.choice(population_data.index, size=sample_size, replace=False)
sample_data = population_data.loc[sample_indices]

# Display the synthetic sample data
print("\nSynthetic Sample Data (Simple Random Sampling):")
print(sample_data.head())
```

# Types of Studies

After establishing methods for data collection, the subsequent step involves determining the type of study, which inherently influences the nature of relationships that can be inferred. Two predominant types of studies exist:

**Observational:**
- In an observational study, researchers record or observe observations or measurements without manipulating any variables.
- These studies may reveal potential relationships, but they do not establish a cause-and-effect relationship.
  
**Experimental:**
- An experimental study involves the random assignment of a treatment, enabling researchers to draw cause-and-effect (causal) conclusions.
- Also known as scientific studies or experiments, these studies employ random assignment, distinguishing them from observational studies.

**Example 1-2: Types of Studies - Quiz and Exam Score Studies:**
Consider a scenario where quizzes are optional in a class:
- In an observational study, the observation may indicate that high-performing students tend to take quizzes, correlating with better exam scores.
- In an experimental study, quizzes would be randomly assigned to specific students to assess whether taking quizzes causally leads to improved exam scores.

**Causation:**
- Distinguishing between observational and experimental studies is crucial, as cause-and-effect conclusions should be approached with skepticism in observational studies.
- Random assignment of treatments, a characteristic of experimental studies, allows for the establishment of cause-and-effect relationships.

**Ethical Considerations:**
- Experimental design involves ethical considerations. For instance, the initial association between smoking and lung cancer was based on an observational study, not an assignment of smoking behavior. Ethical constraints may limit the random assignment of certain treatments.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Simulate a hypothetical dataset for quiz and exam scores
np.random.seed(42)

# Observational Study
students_obs = pd.DataFrame({
    'QuizScores': np.random.normal(loc=70, scale=10, size=100),
    'ExamScores': np.random.normal(loc=75, scale=15, size=100),
})

# Experimental Study
students_exp = pd.DataFrame({
    'ExamScores_Pre': np.random.normal(loc=75, scale=15, size=100),
    'QuizEffect': np.random.normal(loc=5, scale=3, size=100),
})
students_exp['QuizScores_Post'] = students_exp['ExamScores_Pre'] + students_exp['QuizEffect']

# Display data for observational study
print("Observational Study Data:")
print(students_obs.head())

# Display data for experimental study
print("\nExperimental Study Data:")
print(students_exp.head())

# Visualization of Observational Study
plt.scatter(students_obs['QuizScores'], students_obs['ExamScores'], label='Observational Study')
plt.xlabel('Quiz Scores')
plt.ylabel('Exam Scores')
plt.title('Observational Study: Quiz Scores vs Exam Scores')
plt.legend()
plt.show()

# Visualization of Experimental Study
plt.scatter(students_exp['QuizScores_Post'], students_exp['ExamScores_Pre'], label='Experimental Study')
plt.xlabel('Quiz Scores (Post-Treatment)')
plt.ylabel('Exam Scores (Pre-Treatment)')
plt.title('Experimental Study: Quiz Effect on Exam Scores')
plt.legend()
plt.show()

# Perform t-test to compare mean exam scores between observational and experimental groups
t_stat, p_value = ttest_ind(students_obs['ExamScores'], students_exp['ExamScores_Pre'])
print(f"\nT-Test Results:")
print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_value}")
```

# Variables

**Variable:**
A variable is any characteristic, number, or quantity that can be measured, counted, or observed for record.

**Response Variable:**
The response variable is the focus of the researcher's inquiry and is often referred to as the outcome or dependent variable.

**Explanatory Variable:**
Explanatory variables are those that aim to explain changes in the response. They may also be called predictor or independent variables.

**Other Variables:**

**Lurking Variable:**
A lurking variable is neither the explanatory nor the response variable but exhibits a relationship, such as correlation, with both. It is not considered in the study but could influence the relationship between the studied variables.

**Confounding Variable:**
A confounding variable is within the study and is related to other study variables, impacting the relationship between them. If a lurking variable is included in the study and has a confounding effect, it is then classified as a confounding variable.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulate a hypothetical dataset for study hours, coffee consumption, stress, and exam scores
np.random.seed(42)

# Generate synthetic data representing a student study scenario
num_students = 100
study_hours = np.random.normal(loc=20, scale=5, size=num_students)
coffee_consumption = np.random.normal(loc=3, scale=2, size=num_students)
stress = np.random.normal(loc=50, scale=10, size=num_students)  # Lurking variable
exam_scores = 0.5 * study_hours + 1.5 * coffee_consumption - 0.8 * stress + np.random.normal(loc=70, scale=10, size=num_students)

# Create a DataFrame
student_data = pd.DataFrame({
    'StudyHours': study_hours,
    'CoffeeConsumption': coffee_consumption,
    'Stress': stress,
    'ExamScores': exam_scores
})

# Display the synthetic student data
print("Synthetic Student Data:")
print(student_data.head())

# Visualize the relationship between study hours, coffee consumption, and exam scores
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].scatter(student_data['StudyHours'], student_data['ExamScores'])
axs[0, 0].set(xlabel='Study Hours', ylabel='Exam Scores')

axs[0, 1].scatter(student_data['CoffeeConsumption'], student_data['ExamScores'])
axs[0, 1].set(xlabel='Coffee Consumption', ylabel='Exam Scores')

axs[1, 0].scatter(student_data['Stress'], student_data['ExamScores'])
axs[1, 0].set(xlabel='Stress', ylabel='Exam Scores')

axs[1, 1].scatter(student_data['StudyHours'] + student_data['CoffeeConsumption'], student_data['ExamScores'])
axs[1, 1].set(xlabel='Study Hours + Coffee Consumption', ylabel='Exam Scores')

plt.tight_layout()
plt.show()
```

# Experimental Design

Successful experimental design necessitates adherence to several fundamental principles, ensuring that observed differences in experiment outcomes, not reasonably ascribable to chance, are likely attributable to the applied treatments. These principles include:

**1. Control:**
Control involves the careful management of factors beyond the primary areas of interest. By controlling for these extraneous variables, researchers aim to isolate the effects of the treatments under investigation.

**2. Randomization:**
Randomization is the process of randomly assigning subjects to different groups. This approach minimizes the risk of unintentional biases in group selection, enhancing the internal validity of the study. Randomization contributes to the establishment of a causal relationship between the explanatory variable (treatment) and the response variable.

**3. Replication:**
Replication involves employing a sufficient number of subjects to ensure that randomization creates comparable groups. This not only enhances the precision of the study but also increases the likelihood of detecting meaningful differences among treatments when such differences genuinely exist.

*Benefits of Randomization:*
- Causal Conclusions: Random assignment of treatments allows researchers to draw causal or cause-and-effect conclusions. This establishes the treatment as the explanatory variable and the outcome as the response variable.
  
- Generalizability: If subjd likeects are randomly selected from a broader population, results can be extended to that population. When both random assignment and selection are integral to the study, it qualifies as a completely randomized experiment. In the absence of random assignment (e.g., observational study), the treatment can only be associated with the outcome.

Adhering to these principles enhances the internal and external validity of experimental studies, providing a robust foundation for drawing meaningful conclusions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Set a random seed for reproducibility
np.random.seed(42)

# Simulate a randomized experiment with two treatment groups
control_group = np.random.normal(loc=50, scale=10, size=30)
treatment_group = np.random.normal(loc=55, scale=10, size=30)

# Perform a t-test to compare means between control and treatment groups
t_stat, p_value = ttest_ind(control_group, treatment_group)

# Display the results
print("Results of the Randomized Experiment:")
print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_value}")

# Visualization of the experiment
plt.boxplot([control_group, treatment_group], labels=['Control Group', 'Treatment Group'])
plt.ylabel('Outcome Measurement')
plt.title('Randomized Experiment: Control vs Treatment')
plt.show()
```

# Summarize Quantitative Data
Quantitative data takes on numerical values with a measurable distance between them. Quantitative values can be further classified into:

- **Discrete:** Represents "counted" values, such as the number of people in attendance at an event.
- **Continuous:** Represents "measured" values, such as the weight or height of a person.

The key distinction lies in the nature of the values, with qualitative data representing categories and quantitative data involving numerical measures, either discrete or continuous.

## Descriptive Measures of Quantitative Data

### Measures of Central Tendency: Mean, Median, and Mode

Measures of central tendency are essential aspects of quantitative data, providing an estimate of a "typical" value. Three key measures are the mean, median, and mode.

### Mean

The mean ($\bar{x}$) represents the average of a dataset. For a sample of size $n$ with measurements $x_1, x_2, ..., x_n$, the mean is calculated as:

$$ \bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i $$

### Median

The median is the middle value of ordered data. Steps to find the median:

1. Arrange data in increasing order.
2. Find the location of the median by $\frac{n+1}{2}$, where $n$ is the sample size.
3. The value at the determined location is the median.

### Mode

The mode is the most frequently occurring value in the dataset. There may be more than one mode.

**Example 1-5: Test Scores**

Consider the aptitude test scores: 95, 78, 69, 91, 82, 76, 76, 86, 88, 80.

- **Mean:**
$$ \bar{x} = \frac{1}{10} (95 + 78 + 69 + 91 + 82 + 76 + 76 + 86 + 88 + 80) = 82.1 $$

- **Median:**
Ordering the data: 69, 76, 76, 78, 80, 82, 86, 88, 91, 95.
Median location: $\frac{10+1}{2} = 5.5$, median = 81.

- **Mode:**
Mode is 76, the most frequent value.

**Effects of Outliers**

The mean is sensitive to extreme values, while the median is resistant. For example, changing the score 91 to 9 significantly affects the mean but not the median.

### Shape

The shape of data determines the most appropriate measure of central tendency. Shapes include:

- **Symmetric:**
  - Mean = Median = Mode

- **Left-Skewed:**
  - Mean < Median
  - long tail on the left

- **Right-Skewed:**
  - Mean > Median
  -

Knowing the shape helps in choosing robust measures, especially in the case of skewed data.
